### to set the flag in loyaity flag column

```python
# if the max order by customer is more then 40 the customer will be flagged as 'loyal customer' 
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# if the max order by customer is more then 10 and upto 40 the customer will be flagged as 'Regular customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# if the max order by customer is less then 10 the customer will be flagged as 'New customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# getting the stats on this new column
df_ords_prods_merge['loyalty_flag'].value_counts( dropna = False)
```
